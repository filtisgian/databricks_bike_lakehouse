```python
query ="""
SELECT s.order_number,p.product_key,c.customer_key, s.order_date,s.ship_date, s.due_date, s.sales_amount, s.quantity, s.price 
FROM silver.crm_sales s
LEFT JOIN gold.dim_products p
ON s.product_number=p.product_number
LEFT JOIN gold.dim_customers c
ON s.customer_id=c.customer_id
"""
df=spark.sql(query)
display(df)
```
